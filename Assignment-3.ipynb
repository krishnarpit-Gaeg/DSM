**1. When to Use a for Loop**
**//Use a for loop when:**
You know in advance how many times you want to iterate.
You are iterating over a sequence (list, tuple, string, dictionary, set) or any iterable.

# Q2. Write a python program to print the sum and product of the first 10 natural numbers using for and while loop.

```python
# Using for loop
# sum 
result = 0
product = 1
for i in range(1,11):
    result = result + i
    product = product * i
print(result)
print(product)


# using while loop
i = 1
result = 0
product = 1
while i < 11:
    result = result + i
    product = product * i
    i += 1
    
print(result)
print(product)
```

##  Q3. Create a python program to compute the electricity bill for a household.The per-unit charges in rupees are as follows: For the first 100 units, the user will be charged Rs. 4.5 per unit, for the next 100 units, the user will be charged Rs. 6 per unit, and for the next 100 units, the user will be charged Rs. 10 per unit, After 300 units and above the user will be charged Rs. 20 per unit.

## You are required to take the units of electricity consumed in a month from the user as input.Your program must pass this test case: when the unit of electricity consumed by the user in a month is 310, the total electricity bill should be 2250.

```python
units  = int(input("Enter the electricty units: "))
if units <= 100:
    total_bill = units * 4.5
elif units > 100 and units <= 200:
    total_bill = 100*4.5 + (units-100)*6
elif units > 200 and units <= 300:
    total_bill = 100*4.5 + 100*6 + (units - 200) * 10
else:
    total_bill = 100*4.5 + 100*6 + 100*10 + (units - 300)*20
print(int(total_bill))
```

## Q4. Create a list of numbers from 1 to 100. Use for loop and while loop to calculate the cube of each number and if the cube of that number is divisible by 4 or 5 then append that number in a list and print that list.

```python
num_list = list(range(1,101))
print(num_list)
```

```python
#using for loop
l = []
for num in num_list:
    cube = num ** 3
    if cube%4 == 0 or cube%5 == 0:
        l.append(num)
print(l)
```

```python
# using while loop
i = 0
l1 = []
while i < len(num_list):
    cube  = num_list[i] ** 3
    if cube % 4 == 0 or cube % 5 == 0:
        l1.append(num_list[i])
    i += 1
print(l1)
```

# Q5. Write a program to filter count vowels in the below-given string.
## string = "I want to become a data scientist"

```python
vowlas = ["a", "e", "i", "o", "u"]
string = "I want to become a data scientist"
vowal_count = 0
for i in string:
    if i in vowlas:
        print(i)
        vowal_count += 1
vowal_count
```
