# Q1. What is an API? Give an example, where an API is used in real life.

An API (Application Programming Interface) is like a bridge that allows two different software systems to talk to each other. It defines a set of rules and protocols for how one application can request services or data from another.

**Real-Life Example of API**

Google Maps API: When you book a cab on Uber or Ola, the app doesn’t build its own maps. Instead, it uses Google Maps API to fetch routes, traffic info, and estimated travel time.

Payment Gateway API: When you buy something online, the e-commerce site uses APIs from PayPal, Razorpay, or Stripe to securely process your payment.

Weather API: Apps like travel planners or fitness trackers use weather APIs to show live forecasts without building their own weather system.

# Q3. What is a Web API? Differentiate between API and Web API

A Web API is an API that specifically works over the web using HTTP/HTTPS protocols.

It allows applications to communicate with each other through the internet.


API: General interface for communication between two software components (can be local or remote).

Web API: A specific type of API that works over the internet using HTTP/HTTPS.

Scope: API can be used in operating systems, libraries, or databases; Web API is used in web-based applications.

Data Format: API can use any format; Web API usually uses JSON or XML.

Example: API → Python’s os module; Web API → Google Maps API.

# Q4. Explain REST and SOAP Architecture. Mention shortcomings of SOAP.

**REST (Representational State Transfer)**

Style: REST is an architectural style, not a protocol.

Communication: Uses standard HTTP methods → GET, POST, PUT, DELETE.

Data Format: Usually JSON (lightweight), sometimes XML.


**SOAP (Simple Object Access Protocol)**

Style: SOAP is a strict protocol with rules defined by W3C.

Communication: Uses XML messages wrapped in an “envelope.”


# Q5. Differentiate between REST and SOAP.

**REST**

Architectural style (not a strict protocol).

Uses HTTP methods (GET, POST, PUT, DELETE).

Lightweight, usually JSON (sometimes XML).

Stateless, fast, scalable, widely used in modern web apps.

**SOAP**

Strict protocol defined by W3C.

Uses XML messages wrapped in an envelope.

Heavy, complex, but supports strong security (WS-Security).

Can work over multiple protocols (HTTP, SMTP, TCP).
