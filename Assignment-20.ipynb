# Q1. What is Web Scraping? Why is it Used? Give three areas where Web Scraping is used to get data.

Definition: Web scraping is the technique of automatically fetching and parsing data from websites.

**Why it’s used:**

Manual copy-pasting is slow and error-prone.

Scraping automates data collection, making it faster, scalable, and more accurate.

Businesses and researchers gain insights for decision-making.

**Three Key Areas Where Web Scraping is Used**

1. Market & Competitor Analysis
2. Financial Data Collection
3. Social Media Monitoring


# Q2. What are the different methods used for Web Scraping?

1. Manual Copy-Paste
2. HTML Parsing (Requests + BeautifulSoup)
3. DOM Parsing with Browser Automation (Selenium/Playwright)

# Q3. What is Beautiful Soup? Why is it used?

Beautiful Soup is a Python library used for parsing HTML and XML documents.

**Why is it Used?**

Simplifies Web Scraping
Data Extraction
Flexible Searching


# Q4. Why is flask used in this Web Scraping project?

Flask is not directly responsible for scraping data — that’s handled by libraries like Requests and BeautifulSoup.
Instead, Flask acts as the web framework that provides a way to serve, display, and interact with the scraped data through a web application.

# Q5. Write the names of AWS services used in this project. Also, explain the use of each service.

In this web scraping project, several AWS services are typically used to handle scraping, storage, and deployment. **Amazon EC2** is used to run the scraping scripts and the Flask application on virtual servers. **Amazon S3** stores the scraped data such as CSV, JSON, or images in a scalable and durable way. **Amazon RDS** provides a relational database (like MySQL or PostgreSQL) to organize and query the structured scraped data. **AWS Lambda** can be used to run scraping tasks as serverless functions triggered by events, making it efficient for scheduled jobs. **Amazon CloudWatch** monitors the scraping processes, logs errors, and sets alerts for failures. **Amazon API Gateway** exposes the scraped data as APIs so other applications can consume it. Finally, **Elastic Beanstalk** simplifies deployment of the Flask web app, automatically handling scaling, load balancing, and environment setup. Together, these services create a complete pipeline for scraping, storing, monitoring, and serving data.
