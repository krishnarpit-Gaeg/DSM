# Q1. Explain GET and POST methods.

**1.GET Method**

Purpose: Retrieve data from the server without changing it.

Data Location: Sent via URL query parameters (e.g., example.com/page?user=Krishnarpit).

Visibility: Data is visible in the browser’s address bar.

Security: Less secure — never use for passwords or sensitive info.



**2.POST Method**

Purpose: Send data to the server to create or update resources.

Data Location: Sent in the request body, not visible in the URL.

Visibility: Hidden from the address bar and browser history.

# Q2. Why is request used in Flask?

In Flask, the request object is like a messenger that carries all the information sent by the client (browser or API caller) to your server. Without it, your Flask app wouldn’t know what data came in or how to respond properly.

# Q3. Why is redirect() used in Flask?

In Flask, redirect() is used to send the client (browser) to a different URL instead of staying on the current route. Think of it as saying: “Hey browser, don’t stay here — go there!”

# Q4. What are templates in Flask? Why is the render_template() function used?

In Flask, templates are basically HTML files with placeholders that let you dynamically insert Python data into web pages. They make it possible to separate your application logic (Python code) from your presentation layer (HTML/CSS). Flask uses the Jinja2 template engine for this purpose.

**Why render_template() is used**

It connects your Flask route with the HTML page
