# Q1. What is multithreading in python? Why is it used? Name the module used to handle threads in python.

Multithreading in Python means running multiple threads (smaller units of a process) concurrently within a single program. Each thread can execute a part of the code independently, which gives the illusion of parallelism.

Why is it used?

Concurrency: To perform multiple tasks at the same time (e.g., downloading files while processing data).

Responsiveness: Keeps applications responsive (like a GUI that doesn’t freeze while performing background tasks).

Efficiency: Useful for I/O-bound tasks (network requests, file operations) where threads can work while others wait.

Resource Sharing: Threads share memory space, making communication between them faster compared to processes.


**The module used to handle threads in Python is:**  threading

# Q2. Why threading module used? Write the use of the following functions activeCount(), currentThread(), enumerate()

why the threading module is used:

The threading module in Python is used to create and manage threads. It allows programs to run multiple tasks concurrently, which is especially useful for:

I/O-bound tasks (like reading files, network requests) → threads can work while others wait.

Responsiveness → keeps applications (like GUIs) from freezing.

Concurrency → multiple tasks can be handled “at the same time” within one process.

**1. threading.activeCount()**

    Returns the number of currently active threads.
    Use: Helps monitor how many threads are running at a given time.

**2. threading.currentThread()**

    Returns the Thread object representing the thread that is currently executing
    Use: Useful when you want to know which thread is running a particular piece of code.

**3. threading.enumerate()**

    Returns a list of all active Thread objects.
    Use: Helps you inspect or manage all threads currently alive.

# Q3. Explain the following functions run() start() join() isAlive()

**1.run()**

Defines what the thread will do.

Normally, you don’t call it directly — instead, you pass a function to the thread, and start() will internally call run().


**2.start()**

Begins the thread’s activity.

It tells Python: “Create a new thread and execute the run() method in it.”

**3.join(timeout=None)**

Makes the main program wait until the thread finishes.

If you give a timeout, it waits only that many seconds.

**4. isAlive() / is_alive()**

Returns True if the thread is still running, False otherwise.

In modern Python, use is_alive() (with underscore).

# Q4. Write a python program to create two threads. Thread one must print the list of squares and thread two must print the list of cubes.

```python
import threading
```

```python
def print_squares(numbers):
    for n in numbers:
        print(f"Square of {n} = {n*n}")
```

```python
def print_cubes(numbers):
    for n in numbers:
        print(f"Cube of {n} = {n*n*n}")
```

```python
nums = [1, 2, 3, 4, 5]
```

```python
t1 = threading.Thread(target=print_squares, args=(nums,))
t2 = threading.Thread(target=print_cubes, args=(nums,))
```

```python
t1.start()
t2.start()
```

# Q5. State advantages and disadvantages of multithreading.

**Advantage**

Do many tasks together → Like listening to music while downloading files.

Faster for waiting jobs → If one thread is waiting (like reading a file), another can keep working.

Keeps apps smooth → A program won’t “freeze” because background work runs in another thread.

Shares memory → Threads can easily talk to each other since they live in the same process.

**Disadvantage**

Not true parallel for heavy math → Because of Python’s GIL, only one thread runs Python code at a time.

Harder to manage → Threads can clash if they try to use the same data at once.

Need locks → To avoid mistakes, you must use locks, which can slow things down.

Unpredictable order → Threads may run in different orders each time, so debugging is tricky

# Q6. Explain deadlocks and race conditions.

A deadlock happens when two or more threads are waiting for each other forever, and none can continue.

A race condition happens when two or more threads access and modify shared data at the same time, and the final result depends on the unpredictable timing of execution.
