# Q1, Create a vehicle class with an init method having instance variables as name_of_vehicle, max_speed and average_of_vehicle.

```python
class vehicle:

    def __init__(self,name_of_vehicle, max_speed,average_of_vehicle):
        self.name_of_vehicle = name_of_vehicle
        self.max_speed = max_speed
        self.average_of_vehicle = average_of_vehicle
```

# Q2. Create a child class car from the vehicle class created in Que 1, which will inherit the vehicle class. Create a method named seating_capacity which takes capacity as an argument and returns the name of the vehicle and its seating capacity.

```python
class car(vehicle):
    def seating_capacity(self, capacity):
        return f"name of the vehicle is {self.name_of_vehicle} has seating capacity is {capacity}"
```

```python
c = car("SUV", 97, 37)
```

```python
c.seating_capacity(7)
```

# Q3. What is multiple inheritance? Write a python code to demonstrate multiple inheritance.

Multiple inheritance is a feature in object-oriented programming where a class can inherit attributes and methods from more than one parent class. In Python, this means a child class can derive from multiple base classes, gaining access to all their properties and behaviors.

```python
class Engine:
    
    def engine_info(self):
        return "This vehicle has a powerful engine."
```

```python
class Body:
    
    def body_info(self):
        return "This vehicle has a sleek body design."
```

```python
class Car(Engine, Body):
    
    def car_info(self):
        return "This is a car combining engine and body features."
```

```python
car = Car()
```

```python
car.engine_info()
```

```python
car.body_info()
```

```python
car.car_info()
```

# Q4. What are getter and setter in python? Create a class and create a getter and a setter method in this class.

***1.***  Getter methods are used to access (read) the value of a private attribute.

***2.***   Setter methods are used to modify (write/update) the value of a private attribute.

***3.*** In Python, attributes are often made "private" by prefixing them with an underscore (_) or double underscore (__). Getters and setters provide controlled access to these attributes.

***4.*** This is useful for data encapsulation — protecting internal object state and allowing validation when updating values.

```python
class Person:

    def __init__(self, name, age):
        self.__name = name
        self.__age = age

    def get_info(self):
        return self.__name , self.__age

    def set_name(self, name):
        self.__name = name

    def set_age(self, age):
        if age <=0 or age > 100:
            return "age is not valid"
        else:
            self.__age = age
```

```python
Ram = Person("Ram", 25)
```

```python
Ram.get_info()
```

```python
Ram.set_age(-25)
```

```python
Ram.set_age(29)
```

```python
Ram.get_info()
```

# Q5.What is method overriding in python? Write a python code to demonstrate method overriding.

***1.*** Method overriding occurs when a child class provides a specific implementation of a method that is already defined in its parent class.

***2.*** The method name and parameters must be the same in both parent and child.

***3.*** When the method is called on the child class object, Python uses the child’s version (overridden method), not the parent’s.

***4.*** This is a key concept in polymorphism.

```python
class Person:

    def return_info(self):
        return "this function return the person information"
```

```python
class Person1(Person):

    def return_info(self):
        return "this function return the person information form class person1"
```

```python
P = Person1()
```

```python
P.return_info()
```
