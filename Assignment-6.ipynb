# Q3. Explain why the __init__() function is used. Give a suitable example.

The __init__() function in Python is a constructor method used inside classes. It automatically runs whenever you create a new object from that class. Its main purpose is to initialize the object’s attributes with default or user-provided values, so that the object starts off in a well-defined state.


**Why __init__() is used**

**1.** It sets up initial values for attributes.
**2.** It ensures every object created from the class has the necessary data.
**3.** It makes code cleaner by avoiding manual setup after object creation.

```python
class customer :

    def __init__(self,phone_number,address,email_id,customer_id):
        self.phone_number = phone_number
        self.address = address
        self.email_id = email_id
        self.customer_id = customer_id

    def return_customer_details(self):
        return self.customer_id, self.phone_number, self.email_id, self.address
```

```python
Ram  = customer(5484151541,"mumbai road","ram@gamial.com",251)
```

```python
Ram.return_customer_details()
```

```python
Ram.customer_id
```

# Q5. What is inheritance? Give an example for each type of inheritance.

### 1. Single Inheritance

```python
class cars:
    def car_brand(self):
       print("it will dispaly car brand")

class car(cars):
    def car_model(self):
        print("it will dispaly car model name")
```

```python
c = car()
```

```python
c.car_model()
```

```python
c.car_brand()
```

### 2. Multiple Inheritance

```python
class from_1 :
    def dancing(self):
        print("dancing")
        
class from_2 :
    def riding(self):
        print("riding")

class from_3(from_1,from_2):
    def coding(self):
        print("coding")
```

```python
f3 = from_3()
```

```python
f3.coding()
```

```python
f3.riding()
```

```python
f3.dancing()
```

### 3. Multilevel Inheritance

```python
class Grandparent:
    def money(self):
        print("Grandparent: money")

class Parent(Grandparent):
    def house(self):
        print("Parent: House")

class Child(Parent):   # Multilevel inheritance
    def property(self):
        print("Child: Car")
```

```python
c = Child()
```

```python
c.property()
```

```python
c.money()
```

```python
c.house()
```

### 4.Hierarchical Inheritance

```python
class Vehicle:
    def fuel(self):
        print("Runs on fuel")

class Car(Vehicle):
    def wheels(self):
        print("Car has 4 wheels")

class Bike(Vehicle):
    def wheels(self):
        print("Bike has 2 wheels")
```

```python
c = Car()
```

```python
c.wheels()
```

```python
c.fuel()
```

```python
b = Bike()
```

```python
b.wheels()
```

```python
b.fuel()
```

### 5.Hybrid Inheritance

```python
class A:
    def featureA(self):
        print("Feature A")

class B(A):
    def featureB(self):
        print("Feature B")

class C(A):
    def featureC(self):
        print("Feature C")

class D(B, C):   # Hybrid (Multiple + Hierarchical)
    def featureD(self):
        print("Feature D")
```

```python
d = D()
```
