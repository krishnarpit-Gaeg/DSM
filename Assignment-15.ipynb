# Q1. What is a database? Differentiate between SQL and NoSQL databases.

A database is a structured system for storing information so it can be retrieved and manipulated efficiently.

It allows operations like insertion, updating, querying, and deletion of data.

Examples: Banking systems, e-commerce platforms, social media apps all rely on databases.



**SQL Databases**

Relational (data stored in tables with rows & columns).

Fixed schema (structure predefined).

Follow ACID properties → strong consistency.

Best for complex queries and transactions.

Scale vertically (upgrade hardware).

Examples: MySQL, PostgreSQL, Oracle, MS SQL Server.

Use cases: Banking, ERP, CRM, inventory systems.

**NoSQL Databases**

Non‑relational (data stored in documents, key‑value pairs, graphs, wide‑columns).

Flexible or schema‑less (structure can change easily).

Follow BASE model → high availability, eventual consistency.

Best for large‑scale, unstructured or semi‑structured data.

Scale horizontally (add more servers).

Examples: MongoDB, Cassandra, CouchDB, Neo4j.

Use cases: Big data, IoT, real‑time analytics, social media.

# Q2. What is DDL? Explain why CREATE, DROP, ALTER, and TRUNCATE are used with an example.|

DDL (Data Definition Language) is a subset of SQL used to define and manage database structures.

It deals with schemas, tables, indexes, and constraints rather than the actual data inside them.

DDL commands automatically commit (changes are permanent and cannot be rolled back easily).


CREATE → Think of it as building a new house (new table).

DROP → Demolishing the house (table gone forever).

ALTER → Renovating the house (adding/removing rooms = columns).

TRUNCATE → Emptying the house (all furniture = data removed, but house = table still exists).

```python
CREATE TABLE Courses (
    course_id INT PRIMARY KEY,
    course_name VARCHAR(50)
);

ALTER TABLE Courses ADD duration INT;

TRUNCATE TABLE Courses;

DROP TABLE Courses;
```

# Q3. What is DML? Explain INSERT, UPDATE, and DELETE with an example.

DML is a subset of SQL used to manipulate data inside tables.

Unlike DDL (which defines structure), DML works with the actual records.

Common DML commands: INSERT, UPDATE, DELETE, SELECT.

These commands can be rolled back (unlike most DDL commands).

INSERT → Adding a new entry in your school register.

UPDATE → Correcting or changing details in the register.

DELETE → Removing a student’s entry from the register.

```python
# Insert 
INSERT INTO Students (id, name, age) VALUES (2, 'Riya', 19);

# Update
UPDATE Students SET age = 20 WHERE id = 2;

# Delete 
DELETE FROM Students WHERE id = 2;
```

# Q4. What is DQL? Explain SELECT with an example.

DQL is a subset of SQL used to query and fetch data from databases.

The SELECT statement is the primary command in DQL.

Unlike DDL (structure) and DML (data manipulation), DQL is focused on retrieving information.


**SELECT Command**

Purpose: To retrieve data from one or more tables.

**EXAMPLE**

SELECT * FROM Students;

# Q5. Explain Primary Key and Foreign Key.

A Primary Key is a constraint in a table that uniquely identifies each record (row).

**Properties:**

Must be unique (no duplicates allowed).

Cannot be NULL (every row must have a valid identifier).

Can be a single column or a composite key (multiple columns combined).



**Foreign Key**

A Foreign Key is a column (or set of columns) in one table that refers to the Primary Key in another table.

It establishes a relationship between two tables.

Ensures referential integrity (data remains consistent across tables).

# Q6. Write a python code to connect MySQL to python. Explain the cursor() and execute() method.

```python
import mysql.connector


mydb = mysql.connector.connect(
    host="localhost",       
    user="root",            
    password="yourpassword",
 
)
mycursor = mydb.cursor()
mycursor.execute("SHOW TABLES")
for table in mycursor:
    print(table)
mydb.close()
```

**1.cursor()**

Purpose: Creates a cursor object which acts like a pointer to interact with the database.

Think of it as a pen you use to write queries on the database.

Without a cursor, you cannot send SQL commands.

**2.execute()**

Purpose: Sends an SQL query to the database via the cursor.

It can be used for DDL (CREATE, DROP), DML (INSERT, UPDATE, DELETE), and DQL (SELECT) queries.

After execution, results can be fetched using fetchone(), fetchall(), or by iterating over the cursor.
