# Q1. Which function is used to open a file? What are the different modes of opening a file? Explain each mode of file opening.

open() function is used to open a file.

***File Opening Modes in Python***

***1.*** ***Reading a File***

            Mode : "r"
            Meaning : Read

```python
f = open("read_mode.txt", "r")
print(f.read())
f.close()
```

***2.*** ***Writing to a file***

            Mode : "w"
            Meaning: write

```python
f = open("write_mode.txt", "w")
f.write("Hello, world!")
f.close()
```

***3.*** ***Appending to a file***

            Mode:  "a"
            Meaning: append

```python
f = open("append_mode.txt", "a")
f.write("\nNew line added.")
f.close()
```

***4.*** ***Binary mode (e.g., images)***

            Mode: "rb"
            Meaning: Open the file for reading in binary format

```python
f = open("Binary_mode.png", "rb")
content = f.read()
f.close()
```

# Q2. Why close() function is used? Why is it important to close a file?

***Why close() Function is Used***

When you open a file using open(), Python creates a file object and establishes a connection between your program and the file stored on disk.
The close() function is used to terminate that connection once you’re done working with the file.

***Importance of Closing a File***

***1.*** ***Releases System Resources***

    Every open file consumes memory and operating system resources.
    close() frees those resources, preventing memory leaks.

# Q4. Explain the following with python code: read(), readline() and readlines().

***1.read()***

    Meaning: Reads the entire file content (or a specified number of characters if you pass an argument).
    Returns: A single string containing the file’s content.

```python
with open("sample.txt", "w") as f:
    f.write("Line 1\nLine 2\nLine 3")
```

```python
with open("sample.txt", "r") as f:
    content = f.read()
    print(content)
```

***2.readline()***

        Meaning: Reads one line at a time from the file.
        Returns: A string containing that line (including the newline character \n).
        Useful when you want to process the file line by line.

```python
with open("sample.txt", "r") as f:
    line1 = f.readline()
    line2 = f.readline()
    print("First line:", line1)
    print("Second line:", line2)
```

***3.readlines()***

        Meaning: Reads all lines at once and returns them as a list of strings.
        Each element in the list corresponds to one line in the file.

```python
with open("sample.txt", "r") as f:
    lines = f.readlines()
    print(lines)
```

# Q5. Explain why with statement is used with open(). What is the advantage of using with statement and open() together?

***Why with is Used with open()***

    When you use with open(...) as f:, Python automatically takes care of:
    
    1.Opening the file.
    2.Running your block of code (read/write operations).
    3.Closing the file automatically, even if an error occurs inside the block.

***Advantages of Using with open()***

***1.Automatic File Closing***

        No need to explicitly call close().
        Prevents forgetting to close the file, which can cause resource leaks or data loss.


***2.Error Safety***

        If an exception occurs while reading/writing, the file will still be closed properly.
        This makes your code more robust.


***3.Cleaner Code***

        Less boilerplate.
        Easier to read and maintain.

# Q6. Explain the write() and writelines() functions. Give a suitable example.

***1.write()***

    Purpose: Writes a single string into a file.
    Important: It does not add a newline (\n) automatically — you must include it yourself if needed.
    Returns: The number of characters written.

```python
with open("example.txt", "w") as f:
    f.write("I want to become a Data Scientist\n")
    f.write("This is the second line.")
```

```python
f1 = open("example.txt", "r")
f2 = f1.read()
print(f2)
f1.close()
```

***2.writelines()***

    Purpose: Writes a list of strings into a file.
    Important: It does not insert newline characters automatically — each string in the list must already contain \n if you want separate lines.
    Returns: Nothing (just writes the list content)

```python
lines = [
    "Line 1: Python is powerful\n",
    "Line 2: File handling is easy\n",
    "Line 3: writelines() writes lists\n"
]

with open("example2.txt", "w") as f:
    f.writelines(lines)
```

```python
with open("example2.txt", "r") as f:
    contecxt = f.read()
    print(contecxt)
```
