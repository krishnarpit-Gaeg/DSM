# Q1. Create a function which will take a list as an argument and return the product of all the numbers after creating a flat list. Use the below-given list as an argument for your function.
# list1 = [1,2,3,4, [44,55,66, True], False, (34,56,78,89,34), {1,2,3,3,2,1}, {1:34, "key2": [55, 67, 78, 89], 4: (45, 22, 61, 34)}, [56, 'data science'], 'Machine Learning']

# Note: you must extract numeric keys and values of the dictionary also.

```python
list1 = [1,2,3,4, [44,55,66, True], False, (34,56,78,89,34), {1,2,3,3,2,1}, {1:34, "key2": [55, 67, 78, 89], 4: (45,
22, 61, 34)}, [56, 'data science'], 'Machine Learning']
```

```python
def product_all_list_integer(l):

    flat_list1 = []

    for number in l:
        if type(number) == int or type(number) == float:
            flat_list1.append(number)
        elif type(number) == complex:
            flat_list1.append(number.real)
            flat_list1.append(number.imag)
        elif type(number) == bool:
            flat_list1.append(int(number))
        elif type(number) == list or type(number) == tuple or type(number) == set or type(number) == frozenset:
            for i in number:
                if type(i) == int or type(i) == float:
                    flat_list1.append(i)
                elif type(i) == complex:
                    flat_list1.append(i.real)
                    flat_list1.append(i.imag)
                elif type(i) == bool:
                    flat_list1.append(int(i))
                    
        elif type(number) == dict:
            for i in number.items():
                for j in i:
                    if type(j) == int or type(j) == float:
                        flat_list1.append(j)
                    elif type(j) == complex:
                        flat_list1.append(j.real)
                        flat_list1.append(j.imag)
                    elif type(j) == bool:
                        flat_list1.append(int(j))
                    elif type(j) == list or type(j) == tuple or type(j) == set or type(j) == frozenset:
                        for k in j :
                            if type(k) == int or type(k) == float:
                                flat_list1.append(k)
                            elif type(k) == complex:
                                flat_list1.append(k.real)
                                flat_list1.append(k.imag)
                            elif type(k) == bool:
                                flat_list1.append(int(k))

    def product_of_f_l(flat_list1):
        result = 1
        for  i in flat_list1:
            result = result*i
        return result
    return product_of_f_l(flat_list1)
```

```python
print("Product of the flat list element is: ", product_all_list_integer(list1))
```

# Input Sentence: I want to become a Data Scientist.

# Encrypt the above input sentence using the program you just created.

# Note: Convert the given input sentence into lowercase before encrypting. The final output should be lowercase.

```python
def encryptinn_message(mes):
    encr_mes = ""

    for i in mes.lower():
        if i.isalpha():
                encr_mes = encr_mes + chr((ord("a")-(ord(i)-ord("z"))))
            else:
                encr_mes = encr_mes + chr((ord("A")-(ord(i)-ord("Z"))))
        elif i == " ":
            encr_mes += "$"
        else:
            encr_mes = encr_mes + i

    return encr_mes
```
