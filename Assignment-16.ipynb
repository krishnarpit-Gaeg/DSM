# Q1. What is MongoDB? Explain non-relational databases in short. In which scenarios it is preferred to use MongoDB over SQL databases?

MongoDB is a popular NoSQL database that stores data in a flexible, JSON-like format called documents.

Instead of tables and rows (like SQL), MongoDB uses collections (like tables) and documents (like rows).

Each document can have a different structure, making it highly adaptable for dynamic data.

**Non-Relational Databases (Short Explanation)**

Non-relational databases (NoSQL) don’t rely on fixed schemas or strict table relationships.


**When to Prefer MongoDB over SQL**

MongoDB is chosen over SQL databases in scenarios like:

Unstructured or semi-structured data → e.g., logs, sensor data, social media posts.

Rapidly changing schemas → when fields may vary across records.

Big data & scalability needs → horizontal scaling across servers.

Real-time applications → chat apps, recommendation engines, IoT.

Agile development → quick prototyping without worrying about schema design upfront.

# Q2. State and Explain the features of MongoDB.

1. Document-Oriented Storage → Stores data as JSON-like documents instead of rows/columns.

2. Schema-less (Dynamic Schema) → No fixed schema; documents in a collection can have different structures.

3. High Performance → Fast read/write operations with indexing and optimized queries.

4. Horizontal Scalability (Sharding) → Distributes data across multiple servers for handling big datasets.

5. Replication (High Availability) → Maintains multiple copies of data for failover and reliability.

6. Powerful Query Language → Supports rich queries, filters, and aggregation pipelines.

7. Indexing → Provides various types of indexes to speed up searches.

8. Aggregation Framework → Enables complex data processing like grouping and analytics.

9. File Storage (GridFS) → Stores large files (images, videos) by splitting into chunks.

10. Flexibility & Integration → Works seamlessly with modern programming languages and frameworks.

# Q4. Using the database and the collection created in question number 3, write a code to insert one record,and insert many records. Use the find() and find_one() methods to print the inserted record.

```python
import pymongo

client = pymongo.MongoClient("mongodb+srv://<username>:<password>@cluster0.mongodb.net/")

db = client["test_database"]

students = db["student"]

one_student = {"name": "Krishnarpit", "course": "Data Science", "marks": 95}
students.insert_one(one_student)

print("Inserted One Record:")
print(students.find_one({"name": "Krishnarpit"}))   


many_students = [
    {"name": "Priya", "course": "AI", "marks": 88},
    {"name": "Amit", "course": "ML", "marks": 92},
    {"name": "Sneha", "course": "Python", "marks": 85}
]
students.insert_many(many_students)

print("Inserted Many Records:")
for s in students.find():   # find() returns all documents
    print(s)
```

# Q5. Explain how you can use the find() method to query the MongoDB database. Write a simple code to demonstrate this.

The find() method is used to query documents from a collection.

It returns a cursor object (like a pointer) that you can iterate over to access multiple documents.

You can pass a filter condition inside find() to match specific documents.

If no filter is given → it returns all documents in the collection.

```python
import pymongo


client = pymongo.MongoClient("mongodb+srv://<username>:<password>@cluster0.mongodb.net/")


db = client["test_database"]
students = db["students"]

students.insert_many([
    {"name": "Krishnarpit", "course": "Data Science", "marks": 95},
    {"name": "Priya", "course": "AI", "marks": 88},
    {"name": "Amit", "course": "ML", "marks": 92},
    {"name": "Sneha", "course": "Python", "marks": 85}
])


for s in students.find():
    print(s)
```

# Q6. Explain the sort() method. Give an example to demonstrate sorting in MongoDB.

The sort() method is used to arrange documents in ascending or descending order based on one or more fields.

By default:

1 → Ascending order

-1 → Descending order

```python
import pymongo


client = pymongo.MongoClient("mongodb+srv://<username>:<password>@cluster0.mongodb.net/")


db = client["test_database"]
students = db["student"]


students.insert_many([
    {"name": "Krishnarpit", "course": "Data Science", "marks": 95},
    {"name": "Priya", "course": "AI", "marks": 88},
    {"name": "Amit", "course": "ML", "marks": 92},
    {"name": "Sneha", "course": "Python", "marks": 85}
])


print("Students sorted by marks (ascending):")
for s in students.find().sort("marks", 1):
    print(s)


print("\nStudents sorted by marks (descending):")
for s in students.find().sort("marks", -1):
    print(s)
```

# Q7. Explain why delete_one(), delete_many(), and drop() is used.

**1. delete_one()**

Purpose → Deletes only the first matching document from a collection.

Use Case → When you want to remove a single record (like deleting one student’s data).


**2. delete_many()**

Purpose → Deletes all documents that match a given filter.

Use Case → When you want to remove multiple records at once (like deleting all students with marks < 50).


**3. drop()**

Purpose → Deletes the entire collection (including all documents and metadata).

Use Case → When you no longer need the collection at all.
