# Q1. What is an Exception in python? Write the difference between Exceptions and Syntax errors.

**1** An Exception is an error that occurs during program execution (runtime).

**2** It interrupts the normal flow of the program.

**3**  Python raises exceptions when something unexpected happens, like dividing by zero, accessing a missing file, or using an invalid index.

**4** You can handle exceptions using try-except blocks so your program doesn’t crash.

**Difference between Exceptions and Syntax Errors**

**Syntax Error**

When they occur: At compile-time (or when the interpreter first reads your code).

Cause: The code structure itself is invalid — you’ve broken the grammar rules of Python.

Python won’t even run this; it immediately throws a SyntaxError.

Analogy: Like writing a sentence with broken grammar — “He go store.” The teacher stops you before you even finish reading.

Example:
print("Hello"  # Missing closing parenthesis


**Exception Error**

When they occur: At runtime (while the program is executing).

Cause: The code is syntactically correct, but something unexpected happens during execution.

Example:
x = 10 / 0  # Division by zero

This runs fine until it hits the division, then raises a ZeroDivisionError.

Analogy: Like reading a grammatically correct sentence — “He went to the store, but the store was closed.” The grammar is fine, but reality causes a problem.

# Q2. What happens when an exception is not handled? Explain with an example 

If an exception is not handled in Python:

The program terminates immediately at the point of error.

Python displays an error message (traceback) showing where the problem occurred and what type of exception it was.

Any code written after the error line will not execute.

```python
print("Program starts")
result = 10 / 0 
print("Program ends") 
```

```python
print("Program starts")
try:
    result = 10 / 0
except ZeroDivisionError:
    print("Oops! Division by zero error handled.")
```

# Q3. Which Python statements are used to catch and handle exceptions? Explain with an example.

**Python provides a set of statements for exception handling:**

**1.** try → Defines a block of code to test for errors.

**2.** except → Defines a block of code to handle the error if it occurs.

**3.** else → Runs if no exception occurs (optional).

**4.** finally → Runs no matter what (used for cleanup tasks like closing files or releasing resources).

```python
try:
    num = int(input("Enter a number: "))
    result = 10 / num

except ZeroDivisionError:
       print("Error: You cannot divide by zero.")

except ValueError: 
    print("Error: Invalid input. Please enter a number.")

else:
    print("Success! Result is:", result)

finally:
    print("Execution finished.")
```

# Q5. What are Custom Exceptions in python? Why do we need Custom Exceptions? Explain with an example.

Custom Exceptions are user-defined exceptions.

They allow you to create your own error types by extending Python’s built-in Exception class.

This is useful when the built-in exceptions (like ValueError, ZeroDivisionError) don’t clearly describe the specific error situation in your program.

**Why do we need Custom Exceptions?**

To make errors more meaningful and specific to your application.

To improve readability and debugging.

To enforce business rules or constraints that Python doesn’t check by default

```python

class InsufficientFundsError(Exception):
    def __init__(self, balance, amount):
        super().__init__(f"Insufficient funds: Balance = {balance}, Tried to withdraw = {amount}")


def withdraw(balance, amount):
    if amount > balance:
        raise InsufficientFundsError(balance, amount)
    else:
        balance -= amount
        print(f"Withdrawal successful! Remaining balance = {balance}")


try:
    withdraw(1000, 1500)   
except InsufficientFundsError as e:
    print("Custom Exception Caught:", e)
```

# Q6. Create a custom exception class. Use this class to handle an exception.

```python
class NegativeNumberError(Exception):
    def __init__(self, value):
        super().__init__(f"Negative number not allowed: {value}")


def square_root(num):
    if num < 0:
        raise NegativeNumberError(num)  
    else:
        return num ** 0.5
```

```python
try:
    result = square_root(9)   
    print("Square root is:", result)
except NegativeNumberError as e:
    print("Custom Exception Caught:", e)
```
