# Q1. Write a program to accept percentage from the user and display the grade according to the following criteria:

#  Marks	                           Grade
# >90	                                 A
# >80 and <= 90	                         B
# >=60 and <= 80	                     C
# below 60	                             D


```python
percentage = int(input("Enter the percentage: "))
if percentage > 90:
    print("Grade A")
elif percentage > 80 and percentage <= 90:
    print("Grade B")
elif percentage >= 60 and percentage <= 80:
    print("Grade C")
else:
    print("Grade D")
```

# Q4. Check how many times a given number can be divided by 3 before it is less than or equal to 10.

```python
number = int(input("Enter The Number : "))
counter = 0
if number <=10:
    print("please Enter the more Greater Number")
else:
    counter = 1
    while True:
        result =number //3
        if result >=10:
            number = result
            counter+=1
            continue
        else:
            break
    print(counter)
```

# Q5.Why and When to Use while Loop in Python give a detailed description with example.

***A while loop in Python is used to repeatedly execute a block of code as long as a given condition remains true. It’s one of the fundamental #control flow tools in programming.***

**Why Use a while Loop?**
*1*. Unknown number of iterations: When you don’t know beforehand how many times the loop should run.
*2.* Condition-driven repetition: Useful when the loop depends on a dynamic condition (e.g., user input, sensor data, file reading).
*3.* Flexibility: Unlike for loops (which iterate over sequences), while loops can handle more complex stopping conditions.

**When to Use a while Loop**
*1* User input validation: Keep asking until valid input is given.
*2.* Reading data streams: Continue until reaching the end of a file or data source.
*3.* Simulations: Run until a certain threshold or condition is met.
*4.* Infinite loops with break conditions: For programs that should run continuously until explicitly stopped.

```python
i = 0
while i < 7:
    print(i)
    i +=1
```

# Q6. Use nested while loop to print 3 different pattern.

```python
row = 0
col  = 0
while row <= 5:
    col = 0
    while col < row:
        print("*", end  = " ")
        col +=1
    row +=1
    print()
```

```python
i = 5
while i >= 1:  
    j = 1
    while j <= i: 
        print("*", end=" ")
        j += 1
    print()
    i -= 1
```

```python
i = 1
while i <= 5:
    j = 5
    while j > i:
        print(" ", end=" ")
        j -= 1
    
    k = 1
    while k <= (2*i - 1):
        print("*", end=" ")
        k += 1
    
    print()
    i += 1
```
