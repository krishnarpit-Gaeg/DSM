# Q2. Create a simple Flask application to display ‘Hello World!!’. Attach the screenshot of the output in Jupyter Notebook.

# Q3. What is App routing in Flask? Why do we use app routes?

App routing means mapping a URL path (like /home, /about) to a Python function in a Flask application.

When a user visits that URL in the browser, Flask’s routing system calls the corresponding function and returns its output (HTML, text, JSON, etc.).

Technically, this is done using the @app.route() decorator or the app.add_url_rule() method.


**Why Do We Use App Routes?**

1. Connects specific URLs to specific functions, so the app knows what to execute when a user visits a page.

2. Allows multiple pages (/home, /about, /contact) in a single application

# Q4. Create a “/welcome” route to display the welcome message “Welcome to ABC Corporation” and a “/” route to show the following details:
# Company Name: ABC Corporation
# Location: India
# Contact Detail: 999-999-9999

```python
from flask import Flask

app = Flask(__name__)


@app.route("/")
def Company_details():
    return "<h1>Company Name: ABC Corporation</h1> <h2>Location: India</h2> <h2>Contact Detail: 999-999-9999</h2>"

@app.route("/welcome")
def Welcome_mesg():
    return "<h1>Welcome to ABC Corporation</h1>"


if __name__ == "__main__":
    app.run()
```

# Q5. What function is used in Flask for URL Building? Write a Python code to demonstrate the working of the url_for() function.

**Function Used for URL Building in Flask** The function is url_for().



```python
from flask import Flask
from flask import redirect, url_for

app = Flask(__name__)

@app.route("/")
def home():
    return "<h1>Welcome to the Page</h1>"


@app.route("/home")
def redirect_home():
    return redirect(url_for("home"))

if __name__ == "__main__":
    app.run()
```
