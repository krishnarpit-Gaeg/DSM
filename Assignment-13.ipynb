# Q1. Explain why we have to use the Exception class while creating a Custom Exception.

# Note: Here Exception class refers to the base class for all the exceptions.

**1.** **Hierarchy & Consistency**

In Python, all built-in exceptions (like ValueError, TypeError, etc.) inherit from the base class Exception.
    
By inheriting from Exception, your custom exception becomes part of this hierarchy, ensuring it behaves like other exceptions.


**2.** **Catching Mechanism**

The try-except block works by checking the type of the raised error.

If your custom exception doesn’t inherit from Exception, it won’t be caught by a generic except Exception: clause.

# Q3. What errors are defined in the ArithmeticError class? Explain any two with an example.

ArithmeticError is a built-in base class in Python.
It groups together errors that occur during numeric calculations.

**Subclasses of ArithmeticError include:**
    OverflowError
    ZeroDivisionError
    FloatingPointError


**Explanation of Two Common Subclasses**

**1.**  **ZeroDivisionError**
    Raised when you try to divide a number by zero.

```python
try:
    result = 10 / 0
except ZeroDivisionError as e:
    print("Error caught:", e)
```

**2.** **OverflowError**
        Raised when a calculation exceeds the maximum limit for a numeric type.

```python
import math

try:
    result = math.exp(1000)
except OverflowError as e:
    print("Error caught:", e)
```

# Q4. Why LookupError class is used? Explain with an example KeyError and IndexError.

LookupError is a base class in Python for errors that occur when you try to look up something that doesn’t exist.

It groups together exceptions related to invalid lookups in sequences (like lists, tuples, strings) or mappings (like dictionaries).

Subclasses include:

IndexError → invalid index in a sequence.

KeyError → invalid key in a dictionary.

So, LookupError is like a parent category that says: “This error happened because you tried to look up something that isn’t there.”

```python
# Example 1: KeyError
try:
    data = {"name": "Krishnarpit", "age": 21}
    print(data["location"])   
except KeyError as e:
    print("KeyError caught:", e)
```

```python
# Example 2: IndexError
try:
    numbers = [10, 20, 30]
    print(numbers[5])   
except IndexError as e:
    print("IndexError caught:", e)
```

# Q5. Explain ImportError. What is ModuleNotFoundError?

ImportError is raised when an import statement fails.

This happens if:

The module exists but cannot be loaded (maybe due to missing dependencies or circular imports).

You try to import a name (function/class/variable) that doesn’t exist inside a module.

ModuleNotFoundError

ModuleNotFoundError is a subclass of ImportError.

It is raised specifically when the module itself cannot be found.

Introduced in Python 3.6 to make error messages more precise.

```python
try:
    import non_existing_module 
except ModuleNotFoundError as e:
    print("ModuleNotFoundError caught:", e)
```
