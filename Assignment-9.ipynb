# Q1. What is Abstraction in OOps? Explain with an example.

abstraction in Object-Oriented Programming (OOP) is the concept of hiding implementation details and showing only the essential features of an object. It helps reduce complexity and makes code easier to maintain. In Python, abstraction is often implemented using abstract classes and abstract methods (via the abc module).

```python
import abc

class Shape:

    @abc.abstractmethod
    def parameter(self):
        pass

    @abc.abstractmethod
    def area(self):
        pass
```

```python
class circle(Shape):

    def parameter(self, r):
        return 2*3.14*r

    def area(self, r):
        return 3.14*(r**2)
```

```python
area_par_cir = circle()
```

```python
area_par_cir.parameter(7)
```

```python
area_par_cir.area(7)
```

# Q3. What is abc module in python? Why is it used?

The abc module stands for Abstract Base Classes. It provides the infrastructure to define abstract classes in Python. Abstract classes are classes that cannot be instantiated directly and are meant to serve as blueprints for other classes.

Why It Is Used :
1. To enforce that certain methods must be implemented in subclasses.
2. To define a common interface across multiple classes.
3. To improve code clarity and consistency in larger projects.

# Q4. How can we achieve data abstraction?

Data Abstraction means exposing only the essential features of an object while hiding the internal implementation details. In Python, we achieve abstraction mainly through

***Ways to Achieve Data Abstraction:*** 


***1. Abstract Classes and Methods (abc module)***

1.  Define abstract methods in a base class.
2.  Subclasses must implement these methods.
3.  This hides the internal logic and enforces a common interface.


***2. Encapsulation with Access Modifiers***

1. Use private (__variable) and protected (_variable) attributes.
2. Provide controlled access via getter and setter methods.
3. This hides the internal data and exposes only necessary operations.
